```python
from sklearn.model_selection import RandomizedSearchCV
from sklearn.model_selection import cross_val_score
import lightgbm as lgb

params = {
    'task' : ['predict'],
    'boosting': ['gbdt', 'dart' ],
    'objective': ['mean_absolute_error'],
    'n_estimators': [10, 50, 100,  500, 1000, 2000,5000  ],
    'learning_rate':[0.1,  0.05, 0.01, 0.005 ],
   'num_leaves':[ 7, 15, 31  ],
   'max_depth' :[ 5,10,15,25],
   'min_data_in_leaf':[15,25 ],
  'feature_fraction': [ 0.6, 0.8,  0.9, 1],
    'bagging_fraction': [  0.6, 0.8, 1],
    'bagging_freq': [   100, 200, 400  ],   
}

lgb_model = lgb.LGBMRegressor()
rsearch_lbgr = RandomizedSearchCV(estimator=lgb_model, param_distributions=params, 
                                  n_iter=30, cv=7, verbose=1, random_state=42)
rsearch_lbgr.fit(X_train, y_train)
```

```python
import numpy as np
from sklearn.decomposition import PCA
def load_data(path):
    files_list = sorted(os.listdir(path))
    ids, target = files_list[5:]
    embd = files_list[:5]
    embeddings = []
    for name_embd in embd:
        embeddings.append(np.load(path + '/' + name_embd))
    ids = np.load(path + '/' + ids)
    targets = np.load(path + '/' + target)
    return embeddings, ids, targets

train_path = 'train_embeddings'
atom_train = np.array([])
crys_train = np.array([])
y_train = np.array([])
ids_train = np.array([])
base_neighbours = 190
for batch in sorted(os.listdir(train_path)):
    embeddings, ids, targets = load_data(train_path + '/' + batch)
    targets = targets.reshape(targets.shape[0])
    y_train = np.concatenate((targets, y_train))
    ids_train = np.concatenate((ids, ids_train))
    atom_embds = embeddings[:-1]
    crys_embd = embeddings[-1]
    embeddings = np.array([])
    for atom in atom_embds:
        if atom.shape[0] < base_neighbours:
            n = base_neighbours - atom.shape[0]
            for i in range(n):
                tmp = np.zeros((1, 64))
                atom = np.vstack((atom, tmp))
        if atom.shape[0] > base_neighbours:
            n = atom.shape[0] - base_neighbours
            atom = atom[:-n]
        atom = np.ravel(atom).reshape(1, -1)
        embeddings = np.hstack([atom, embeddings]) if embeddings.size else atom

    atom_train = np.vstack([embeddings, atom_train]) if atom_train.size else embeddings
    crys_train = np.vstack([crys_embd, crys_train]) if crys_train.size else crys_embd

print(atom_train.shape)
print(crys_train.shape)
print(y_train.shape)
print(ids_train.shape)
```

```python
pca1 = PCA(n_components=64)
atom_reduce_train = pca1.fit_transform(atom_train)
```

```python
X_train = np.hstack((atom_reduce_train, crys_train))
print(X_train.shape)
```

```python
pca2 = PCA(n_components=64)
X_train = pca2.fit_transform(X_train)
```
